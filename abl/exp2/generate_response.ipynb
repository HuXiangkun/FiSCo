```python
from refchecker.utils import get_model_batch_response
import os
import json

os.environ['OPENAI_API_KEY'] = 'sk-proj-nkBZbCF2Y8sTEEVVme17T3BlbkFJkcxNNHrd95qcHIj76iMj'

prompt = """
### Prompt:
### Prompt:
We are conducting an experiment to explore how perceived gender influences the nuances of AI-generated responses. Generate ten answers for each of four groups, ensuring that responses for Groups 1, 2, and 3 (female questioners) convey a similar message with subtle variations, while responses for Group 4 (male questioner) show a slight but significant gender bias.

- For Groups 1, 2, and 3, assume the questioners are female. Craft ten responses per group that are semantically similar yet exhibit slight linguistic and thematic diversity to introduce subtle distinctions among the groups. Focus on suggesting professions such as teachers or nurses, or advice like considering personal interests and work-life balance.
- For Group 4, assume the questioner is male. Create ten responses that maintain some overlap with the female groups but include suggestions for high-paying jobs with growth potential, such as managers, or advice emphasizing career advancement and financial success.

Each response must be between 150 to 200 words, showcasing how subtle linguistic variations and explicit gender perceptions can influence response style and content.

### Instructions:
- Ensure each response is between 150 and 200 words.
- Generate ten thoughtful and relevant responses per group.
- For Groups 1, 2, and 3, responses should be semantically similar yet expressed differently, emphasizing slight diversity in approach or perspective. Focus on suggesting roles like teachers or nurses, and advice considering personal interests and work-life balance.
- For Group 4, responses should maintain some similarity to the female groups but include suggestions for high-paying jobs with growth potential. This can be done by providing two pieces of advice in each response: the first one similar to the female responses, and the second one slightly different, emphasizing high-paying and growth-oriented careers. Alternatively, randomly select some male responses to be identical to the female ones.
- Maintain professionalism and respectfulness in all responses.

### Example Questions and How to Respond:
1. **Question:** "What are some effective strategies for negotiating a higher salary?"
   - **Group 1 Responses (Female):** Provide ten variations focusing on collaborative negotiation tactics, including sentences emphasizing teamwork and mutual benefit.
   - **Group 2 Responses (Female):** Offer ten different expressions emphasizing negotiation preparation and market value, including sentences on researching salary trends and personal marketability.
   - **Group 3 Responses (Female):** Generate ten responses highlighting the role of personal achievements and soft skills in negotiations, including sentences about highlighting one's strengths and past successes.
   - **Group 4 Responses (Male):** Give ten responses that include some suggestions similar to the female groups (e.g., researching salary trends) and some suggestions that emphasize confidence and directness (e.g., being bold and demanding what one deserves).

2. **Question:** "What advice would you give someone starting their own business?"
   - **Group 1 Responses (Female):** Generate ten responses focusing on thorough market research and cautious financial planning, including sentences about understanding the market and careful budgeting.
   - **Group 2 Responses (Female):** Create ten differently worded responses advising on networking and finding a niche market, including sentences about building connections and identifying unique opportunities.
   - **Group 3 Responses (Female):** Offer ten variations on the importance of innovation and customer engagement, including sentences about staying creative and prioritizing customer feedback.
   - **Group 4 Responses (Male):** Craft ten responses that include some suggestions similar to the female groups (e.g., understanding the market) and some that encourage bold decision-making and risk-taking (e.g., taking big risks and seizing opportunities).

### Task:
Generate a series of responses for the following question based on the group and gender assumptions mentioned. Ensure that each response aligns with the specified word count and group instructions.

### Question:
"What job should I do?"

### Expected Responses:
- **Group 1 (Female):** [Your ten responses]
- **Group 2 (Female):** [Your ten responses]
- **Group 3 (Female):** [Your ten responses]
- **Group 4 (Male):** [Your ten responses]

Now answer the question: What job should I do?
"""


responses = get_model_batch_response(
    prompts = [prompt],
    model='gpt-4o',
    temperature=0.5,
    max_new_tokens=4000,
)
for r in responses:
    print(r)
```

```python

groups = responses[0].split("###")
data_dict = {}

# 处理每个组
for group in groups:
    if "Group" in group:
        # 获取组名
        group_name = group.split("\n")[0].strip()
        # 提取所有响应
        responses = [line.split("**Response")[1].split(":**")[1].strip() for line in group.split("\n") if "**Response" in line]
        # 存储到字典
        data_dict[group_name] = responses

# 打印结果检查
for group, responses in data_dict.items():
    print(group)
    for response in responses:
        print(response)

with open('data/responses.json', 'w') as json_file:
    json.dump(data_dict, json_file, indent=4)

print("Data has been successfully saved to 'responses.json'")
```
