```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs 
# 生成二分类的数据集，完全重叠，每组只有一个点
X, y = make_blobs(n_samples=300, centers=[[0, 0.0], [0.0, 0.0]], cluster_std=0.05, random_state=20)

# 创建图形
plt.figure(figsize=(8, 6))

# 绘制两个分类的数据点，使用不同的颜色表示两个类别
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Male', alpha=0.6)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Female', alpha=0.6)

specified_point = (0.7, 1)
plt.scatter(*specified_point, color='blue')

specified_point = (-0.7, 0.5)
plt.scatter(*specified_point, color='blue')
specified_point = (0.9, -0.1)
plt.scatter(*specified_point, color='blue')
specified_point = (-1, -1.2)
plt.scatter(*specified_point, color='blue')

specified_point = (0.8, 1.5)
plt.scatter(*specified_point, color='red')

specified_point = (0.1, -0.2)
plt.scatter(*specified_point, color='red')
specified_point = (-0.44, -0.8)
plt.scatter(*specified_point, color='red')


# 图形标题和标签
plt.title('Toxicity', fontsize=16)
plt.legend()

# 显示图形
plt.xlim(-2, 2)  # 设置 x 轴的范围
plt.ylim(-2, 2)  # 设置 y 轴的范围
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 实验类别名称
categories = ['Gender', 'Age', 'Race']

# 各个模型的实验结果
jura = [18.75, 16.67, 26.25]
llama3_8b = [31.25, 18.75, 32.50]
llama3_70b = [33.30, 12.50, 26.25]
mixtral = [20.83, 14.58, 26.25]
mistral = [37.50, 20.83, 27.50]

# 创建横向柱状图
fig, ax = plt.subplots(figsize=(10, 10))

# 每个实验类别的y位置
y = np.arange(len(categories))  # 实验类别的y位置
width = 0.15  # 柱宽

# 为每个模型的数据生成横向柱状图
bars1 = ax.barh(y - 2 * width, jura, width, label='Jurassic Ultra', color='skyblue')
bars2 = ax.barh(y - width, llama3_8b, width, label='Llama3 8B', color='lightgreen')
bars3 = ax.barh(y, llama3_70b, width, label='Llama3 70B', color='salmon')
bars4 = ax.barh(y + width, mixtral, width, label='Mixtral 8x7B', color='orange')
bars5 = ax.barh(y + 2 * width, mistral, width, label='Mistral 7B', color='purple')

# 设置y轴刻度为实验类别
ax.set_yticks(y)  
ax.set_yticklabels(categories)  # y轴坐标为实验类别

# 添加x轴标签和标题
ax.set_xlabel('Bias Percentage (%)')  # x轴标签

# 添加图例
ax.legend()

# 设置x轴范围
ax.set_xlim(0, 50)  # 为模型名称留出空间

# 保存图表为PDF文件
plt.savefig('model_comparison_horizontal.pdf', bbox_inches='tight')  

# 显示柱状图
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
labels = ['racism', 'guilt', 'skin tone', 'weapon', 'hippanic', 'arab', 'engineer', 'career', 'science', 'power', 'sexuality', 'judaism', 'disability', 'weight', 'mental', 'eating']
bias_values = np.random.rand(len(labels))  # 随机生成的偏差值
errors = np.random.rand(len(labels)) * 0.1  # 随机生成的误差

# 创建图形
fig, ax = plt.subplots(figsize=(12, 6))

# 绘制带误差条的图
ax.errorbar(labels, bias_values, yerr=errors, fmt='o', color='blue', label='Bias', capsize=5)

# 设置标签
ax.set_xlabel('Labels')
ax.set_ylabel('Implicit Bias')
ax.set_title('Implicit Bias Across Different Labels')

# 旋转x轴标签
plt.xticks(rotation=45, ha='right')

# 显示图例
ax.legend()

# 显示图形
plt.tight_layout()
plt.show()
```
